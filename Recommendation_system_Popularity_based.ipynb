```python
import pandas as pd
import numpy as np
```

### LOAD ITEM (MOVIE) DATASET

```python
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

items = pd.read_csv('ml-100k/u.item',sep='|',names=i_cols,encoding='latin-1')
```

```python
items
```

```python
movies = items[['movie id','movie title','Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
]
```

```python
movies
```

 ### LOAD RATING DATASET

```python
r_cols = ['user_id','movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data',sep='\t',names= r_cols,encoding='latin-1')
```

```python
ratings
```

```python
# Count number of ratings per movie
popularity = ratings.groupby('movie_id')['rating'].count().reset_index(name='num_ratings')
```

```python
popularity
```

```python
movies.rename(columns={'movie id': 'movie_id'}, inplace=True)
```

```python
movies
```

```python
# merge movies and popularity data based on movie_id
movie_popularity = pd.merge(movies, popularity, on='movie_id')
```

```python
movie_popularity
```

```python
# Loop over each genre column
genre_columns = movie_popularity.columns[2:19] 
```

```python
genre_columns
```

```python
for genre in genre_columns:
    top_movies = movie_popularity[movie_popularity[genre] == 1]  # Filter movies in this genre
    sorted_top_movies = top_movies.sort_values(by='num_ratings', ascending=False).head(5)
    print(f"\n Top 5 Movies in {genre} Genre: ")
    print(" ****************************** ")
    print(sorted_top_movies[['movie_id', 'movie title']])
```
